# Model Creation

```python
#Step 1 Independent and dependent variable creation
X=data[['youtube','facebook','newspaper']]   # Select features (independent variables) for the model
Y=data.sales                                 # Select target variable (dependent variable) — sales
```

```python
X     # Displays the feature matrix containing columns: youtube, facebook, and newspaper
```

```python
Y     # Displays the target variable (sales) as a Series
```

```python
# Step 2 Traning and testing data creation
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
x_train       # Displays the training features after train-test split
```

```python
y_train    # Displays the target values (sales) for the training set after train-test split
```

```python
# Step 3 Creating model
from sklearn.linear_model import LinearRegression   # Import Linear Regression model from scikit-learn
```

```python
lr1=LinearRegression()        # object creation
```

```python
lr1.fit(x_train,y_train)      # training of linear regression
```

```python
y_pred=lr1.predict(x_test)    # Predict sales using the trained model on the test set features
y_pred                        # Displays predicted sales
```

```python
import warnings     # Suppress all warnings 
warnings.filterwarnings("ignore")
```

```python
lr1.predict([[200,150,50]])   # Predict sales for a custom input: 200 (YouTube), 150 (Facebook), 50 (Newspaper)
```

y_test is the ground truth for the test set.
y_pred is the model's prediction for the same test inputs.

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
r2_score=r2_score(y_test,y_pred)
r2_score
```

```python
from sklearn.metrics import mean_squared_error  # Import function to calculate Mean Squared Error (MSE) for model evaluation
mean_squared_error(y_test,y_pred)               # Calculates the Mean Squared Error between actual and predicted sales values 
```

```python
mean_squared_error(y_test,y_pred)**0.5        # Calculates Root Mean Squared Error to measure average prediction error in original units
```
